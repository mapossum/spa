```python
aprx = arcpy.mp.ArcGISProject("CURRENT")
dtmap = aprx.listMaps()[0]
treelayer = dtmap.listLayers()[0]
treelayer.definitionQuery = "SPECIES LIKE '%OAK%'"
aprx.save()
```

```python
aprx = arcpy.mp.ArcGISProject("CURRENT")
dtmap = aprx.listMaps()[0]
treelayer = dtmap.listLayers()[0]
sym = treelayer.symbology

green = {"RGB": [0, 100, 0, 100]}

sym.renderer.symbol.color = green

treelayer.symbology = sym

aprx.save()
```

```python
aprx = arcpy.mp.ArcGISProject("CURRENT")
dtmap = aprx.listMaps()[0]
treelayer = dtmap.listLayers()[0]
sym = treelayer.symbology

colorRamp = aprx.listColorRamps("Greens")[0]
sym.renderer.colorRamp = colorRamp

treelayer.symbology = sym

aprx.save()
```

```python
cr = aprx.listColorRamps()

for c in cr:
    print(c.name)
```

```python
aprx = arcpy.mp.ArcGISProject("CURRENT")
dtmap = aprx.listMaps()[0]
parcellayer = dtmap.listLayers()[2]

sym = parcellayer.symbology
sym.updateRenderer("GraduatedColorsRenderer")
sym.renderer.classificationField = "Shape_Area"
sym.renderer.classificationMethod = "Quantile"
sym.renderer.breakCount = 5
colorRamp = aprx.listColorRamps("Yellow to Red")[0]
sym.renderer.colorRamp = colorRamp

parcellayer.symbology = sym
aprx.save()
```
