```python
import matplotlib.pyplot as plt

def plot_parcel(x_start, y_start, side_length):
    """
    Calculates the 5 coordinates of a closed square and plots them.
    
    x_start, y_start: The bottom-left corner of the parcel
    side_length: How wide/tall the parcel is
    """
    # 1. Define the X coordinates for the 4 corners, returning to start
    # Order: Bottom-Left, Bottom-Right, Top-Right, Top-Left, Bottom-Left
    x_coords = [
        x_start, 
        x_start + side_length, 
        x_start + side_length, 
        x_start, 
        x_start
    ]
    
    # 2. Define the Y coordinates matching the order above
    y_coords = [
        y_start, 
        y_start, 
        y_start + side_length, 
        y_start + side_length, 
        y_start
    ]
    
    # 3. Plot the line connecting these points
    # 'k-' means black color ('k') with a solid line ('-')
    plt.plot(x_coords, y_coords, 'b-') 

# --- Main Execution ---

# Setup the plotting figure
plt.figure(figsize=(8, 4))
plt.title("Survey Row: Coordinate Lists")

# Variable to define parcel size
parcel_size = 20

# Loop to draw 5 parcels
for i in range(5):
    # Calculate the shifting x-coordinate
    current_x = i * parcel_size
    
    # Call the function
    plot_parcel(current_x, 0, parcel_size)

# Visual cleanup
plt.axis('equal')  # Ensures the squares look square
plt.grid(True)     # Adds a grid for reference
plt.show()
```
