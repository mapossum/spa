```python
#This is a script that will disagg a field by another field after an intersect.

import arcpy

arcpy.env.workspace = r"spaWeek06.gdb"

# Inputs
# Feature Class to be disagged
disaggFC = "ctracts"

# The field in the disagg dataset that will be the data that needs to be disaggregated then re-aggregated
oPopFieldName = "P0010001"

# Feature Class to be used as the summarizing area.
overlayFC = "watersheds"
# The field in the overlay FC that represents the unique id for the final summaries
oUnitsfield = "HUC_8"


# Non-inputs that will be used in processing
# InitialArea field setup (This will hold the area of disagg dataset before it is intersected with the summaring  
iaFieldName = "Inital_Area"

# Temporary dataset to store the intersection of the two daatasets
tintersectFC = r"memory\intersect"


# Outputs
EstPopFieldName = "Estimated_population"
disaggoutFC = r"disaggdata"


# Derived name of aggregation field (sum)
TotalPopFieldName = "{} SUM".format(EstPopFieldName)

arcpy.management.CalculateField(disaggFC, iaFieldName, "!shape.area!", "PYTHON3", '', "DOUBLE")
arcpy.analysis.PairwiseIntersect("{};{}".format(disaggFC,overlayFC), tintersectFC, "ALL", None, "INPUT")
arcpy.management.CalculateField(tintersectFC, EstPopFieldName, "(!shape.area! / !{}!) * !{}!".format(iaFieldName, oPopFieldName), "PYTHON3", '', "LONG")
arcpy.analysis.PairwiseDissolve(tintersectFC, disaggoutFC, oUnitsfield, TotalPopFieldName, "MULTI_PART")
```
