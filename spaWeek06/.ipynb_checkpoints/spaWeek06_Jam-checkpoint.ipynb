# Areal Disaggregation and Reaggregation: Estimating Population by Watershed

## Today's Main Task - The Big Picture

One of the most common challenges in GIS is that the data we have and the boundaries we need rarely line up. Census data is collected by census tracts, but our analysis might require population estimates for watersheds, school districts, legislative districts, or any other set of polygons that do not share boundaries with the census geography. This mismatch between **source zones** (where the data lives) and **target zones** (where we need the data) is sometimes part of what is called the **Modifiable Areal Unit Problem (MAUP)**.

### What Is Areal Disaggregation?

**Areal disaggregation** is the process of breaking apart a value that is reported for a larger (or differently shaped) polygon and redistributing it into smaller pieces based on how much area each piece contributes. Once we have those smaller pieces, we can **reaggregate** (sum) them into our target zones.

The general workflow looks like this:

1. Record the original area of each source polygon (census tracts).
2. **Intersect** the source polygons with the target polygons (watersheds). This splits every census tract along every watershed boundary, producing many small slivers.
3. For each sliver, estimate its share of the original value: **(sliver area / original tract area) × original population**.
4. **Dissolve** (reaggregate) those slivers by the target zone identifier, summing the estimated values.

### Why Are We Doing This Here?

Look at the map you currently have open. The **watershed boundaries** (thick black outlines) represent HUC-8 hydrologic units from the USGS National Hydrography Dataset. The **census tracts** are displayed with graduated colors showing total population (`P0010001`). Notice how the two sets of boundaries cut across each other — a single census tract may fall inside two or more watersheds, and a single watershed may contain parts of dozens of census tracts.

Our goal is straightforward: **estimate how many people live in each HUC-8 watershed**. There is no census product that reports population by watershed, so we must derive it ourselves.

### Weaknesses and Assumptions

This approach assumes that population is **uniformly distributed** across the area of each census tract. That assumption is almost never perfectly true — people tend to cluster along roads, in towns, and away from steep slopes or water bodies. As a result, our estimates will contain error, especially for large rural tracts where population is concentrated in a small part of the polygon.

**Possible ways to improve the estimate:**

- **Dasymetric mapping**: Use ancillary data (land cover, building footprints, road networks) to create weights that redistribute population toward areas that are more likely to be inhabited. For example, exclude forested or water-covered areas from the calculation so that population is allocated only to developed land.
- **Use smaller source zones**: Census blocks are much smaller than census tracts and reduce the uniform-distribution error.
- **Use address-level data**: Geocoded address points give you the actual locations of residences, eliminating the need for areal disaggregation entirely.

Even with its limitations, simple area-weighted disaggregation is a powerful and widely used technique. Understanding it is foundational before moving on to more sophisticated methods.

## Data Validation: Checking Geometry Type, Spatial Reference, and Extent Overlap

Before running any geoprocessing, it is good practice (though NOT required) to **validate your inputs**. Many geoprocessing errors can be traced back to one of three common problems:

1. **Wrong geometry type** — An intersect of two polygon layers produces polygons. If one input were a line or point layer, you would get unexpected results or an error.
2. **Mismatched coordinate systems** — If the two layers are in different spatial references (projections), area calculations will be unreliable and the intersect may produce incorrect geometry. ArcGIS Pro will often project data on the fly for display, so the map may *look* correct even when the underlying data is not aligned.
3. **Non-overlapping extents** — If the two datasets do not geographically overlap, the intersect will produce an empty result.

We will write a **function** that performs all three checks and prints clear warnings if anything is wrong. Wrapping these checks in a function means we can reuse the same validation logic in future scripts.

> **Programming concept — functions:** A function is a named, reusable block of code. You define it once with `def`, then call it whenever you need it. Functions reduce code duplication and make scripts easier to read, test, and maintain.

````python
# Cell 6 — Define a data-validation function

def validate_inputs(fc1, fc2):
    """
    Validate that two feature classes are both polygons, share the same
    spatial reference, and have overlapping geographic extents.

    Parameters
    ----------
    fc1 : str  — Path or name of the first feature class.
    fc2 : str  — Path or name of the second feature class.

    Returns
    -------
    bool — True if all checks pass, False otherwise.
    """
    all_passed = True

    # Use arcpy.Describe to get metadata about each feature class
    desc1 = arcpy.Describe(fc1)
    desc2 = arcpy.Describe(fc2)

    # --- Check 1: Geometry type ---
    print("--- Geometry Type Check ---")
    print("  {} : {}".format(fc1, desc1.shapeType))
    print("  {} : {}".format(fc2, desc2.shapeType))
    if desc1.shapeType != "Polygon" or desc2.shapeType != "Polygon":
        print("  WARNING: Both feature classes must be Polygon type for areal disaggregation!")
        all_passed = False
    else:
        print("  PASSED — Both are Polygon feature classes.")

    # --- Check 2: Spatial reference ---
    sr1 = desc1.spatialReference
    sr2 = desc2.spatialReference
    print("\n--- Spatial Reference Check ---")
    print("  {} : {}".format(fc1, sr1.name))
    print("  {} : {}".format(fc2, sr2.name))
    if sr1.name != sr2.name:
        print("  WARNING: The spatial references do not match!")
        print("  Area calculations will be unreliable. Reproject one dataset to match the other.")
        all_passed = False
    else:
        print("  PASSED — Both use the same spatial reference.")

    # --- Check 3: Extent overlap ---
    ext1 = desc1.extent
    ext2 = desc2.extent
    print("\n--- Extent Overlap Check ---")
    print("  {} extent: XMin={:.2f}, YMin={:.2f}, XMax={:.2f}, YMax={:.2f}".format(
        fc1, ext1.XMin, ext1.YMin, ext1.XMax, ext1.YMax))
    print("  {} extent: XMin={:.2f}, YMin={:.2f}, XMax={:.2f}, YMax={:.2f}".format(
        fc2, ext2.XMin, ext2.YMin, ext2.XMax, ext2.YMax))

    # Two rectangles overlap if neither is completely to the left, right,
    # above, or below the other.
    overlaps = not (ext1.XMax < ext2.XMin or ext1.XMin > ext2.XMax or
                    ext1.YMax < ext2.YMin or ext1.YMin > ext2.YMax)
    if not overlaps:
        print("  WARNING: The extents do NOT overlap! The intersect will produce no features.")
        all_passed = False
    else:
        print("  PASSED — The extents overlap.")

    # --- Summary ---
    print("\n=== Validation {} ===".format("PASSED" if all_passed else "FAILED — review warnings above"))
    return all_passed
````

## Running the Validation Function

Now we **call** the function we just defined, passing in our two input feature classes. If any check fails, the function prints a warning and returns `False`. We store the result so we could, in a production script, halt execution if validation fails.

Take a moment to read the output carefully. Confirm that both datasets are polygons, that they share the same coordinate system, and that their extents overlap.

````python
# Cell 7 — Run the validation function on our inputs

is_valid = validate_inputs(disaggFC, overlayFC)

if not is_valid:
    print("\nOne or more validation checks failed.")
    print("Fix the issues above before continuing.")
````

## Quick Review: What Are Geoprocessing Calls?

Every time we write something like `arcpy.management.CalculateField(...)` or `arcpy.analysis.PairwiseIntersect(...)`, we are telling Python to **run a geoprocessing tool** — the same tool you could run from the ArcGIS Pro toolbox GUI.

The general pattern is:

```python
arcpy.<toolbox_alias>.<ToolName>(parameter1, parameter2, ...)
```

- `arcpy.analysis` → Analysis toolbox

Running tools from Python instead of the GUI gives us **reproducibility** (the script documents exactly what was done), **automation** (we can chain many tools together), and **parameterization** (we can swap inputs by changing a variable).

When a geoprocessing tool runs, ArcGIS Pro processes the data just as it would if you had opened the tool dialog and clicked Run. The tool writes messages to the geoprocessing history, and we can capture those messages in Python with `arcpy.GetMessages()`.

## Step 1: Calculate the Original Area of Each Census Tract

Before we split the census tracts, we need to **record how large each tract is right now**. We will add a field called `Initial_Area` and populate it with each tract's area using the `Calculate Field` tool.

**Why is this necessary?** After the intersect, each tract polygon will be cut into slivers. Those slivers will have their own (smaller) areas, and we will have lost the information about how big the original tract was. Without that denominator, we cannot compute the ratio `sliver area / original area`.

**What does `!shape.area!` mean?** In an arcpy field calculation, `!fieldname!` is the syntax for referring to a field. `shape` is a special built-in field that holds each feature's geometry. `shape.area` returns the area of that geometry in the units of the feature class's coordinate system. The exclamation marks tell the Calculate Field tool to evaluate this as a Python expression.

**Tool reference:** [Calculate Field (Data Management)](https://pro.arcgis.com/en/pro-app/latest/tool-reference/data-management/calculate-field.htm)

````python
# Cell 9 — Calculate the initial area of each census tract

# arcpy.management.CalculateField(in_table, field_name, expression, expression_type, code_block, field_type)
#   in_table       : the feature class to modify
#   field_name     : the field to calculate (created automatically if it does not exist when a field_type is provided)
#   expression     : the Python expression to evaluate for each row
#   expression_type: "PYTHON3" for Python 3 expressions
#   code_block     : optional code block for complex expressions (empty string here)
#   field_type     : data type of the field if it needs to be created

arcpy.management.CalculateField(
    in_table=disaggFC,
    field=iaFieldName,
    expression="!shape.area!",
    expression_type="PYTHON3",
    code_block="",
    field_type="DOUBLE"
)

print("Added and calculated '{}' field on '{}'".format(iaFieldName, disaggFC))
print(arcpy.GetMessages())
````

## Step 2: Pairwise Intersect — Splitting Tracts by Watershed Boundaries

The **Pairwise Intersect** tool computes the geometric intersection of two polygon feature classes. Every polygon in the output represents an area where a census tract and a watershed **overlap**. If a census tract falls entirely within one watershed, it appears in the output as a single polygon. If it spans two watersheds, it is split into two polygons — one for each watershed.

The output retains the **attributes from both inputs**, so each sliver carries both the census tract fields (including population and `Initial_Area`) and the watershed fields (including `HUC_8`).

We write the result to the in-memory workspace (`memory\intersect`) to keep things fast and avoid cluttering the geodatabase with temporary data.

**Why Pairwise Intersect instead of Intersect?** `PairwiseIntersect` uses a different computational algorithm that is often faster for large datasets and handles complex geometry more robustly. For most purposes the two tools produce identical results.

**Tool reference:** [Pairwise Intersect (Analysis)](https://pro.arcgis.com/en/pro-app/latest/tool-reference/analysis/pairwise-intersect.htm)

````python
# Cell 11 — Run Pairwise Intersect

# arcpy.analysis.PairwiseIntersect(in_features, out_feature_class, join_attributes, cluster_tolerance, output_type)
#   in_features       : semicolon-separated list of input feature classes
#   out_feature_class : where to write the result
#   join_attributes   : "ALL" keeps all fields from both inputs
#   cluster_tolerance : None uses the default xy tolerance
#   output_type       : "INPUT" preserves the input geometry type (polygon)

arcpy.analysis.PairwiseIntersect(
    in_features="{};{}".format(disaggFC, overlayFC),
    out_feature_class=tintersectFC,
    join_attributes="ALL",
    cluster_tolerance=None,
    output_type="INPUT"
)

intersect_count = arcpy.management.GetCount(tintersectFC)
print("Pairwise Intersect complete.")
print("Output features: {}".format(intersect_count))
print(arcpy.GetMessages())
````

Zoom in to an area where a watershed boundary crosses through several census tracts. Notice how each original tract is now split into pieces, and each piece inherits the attributes of both the tract it came from and the watershed it falls within.

We also list the fields in the intersect output so you can verify that both the tract fields and the watershed fields are present.

````python
# Cell 12 — Add the intersect result to the map and list its fields 
# (it is probably already there if you are running this in ArcGIS Pro)

# List all field names in the intersect output
print("\nFields in the intersect output:")
for field in arcpy.ListFields(tintersectFC):
    print("  {:<30} {}".format(field.name, field.type))
````

## Step 3: Calculate the Estimated Population for Each Sliver

This is the heart of the disaggregation. For each sliver polygon in the intersect output, we calculate:

$$\text{Estimated Population} = \frac{\text{Sliver Area}}{\text{Original Tract Area}} \times \text{Original Tract Population}$$

- `!shape.area!` gives the **current** area of the sliver (after the intersect).
- `!Initial_Area!` gives the area of the **original** census tract (calculated in Step 1, before the intersect split it).
- `!P0010001!` gives the **total population** of the original census tract.

The ratio `sliver area / original area` tells us **what fraction of the original tract this sliver represents**. Multiplying by the population gives us the estimated number of people in that sliver.

**Example:** If a census tract had an area of 10 km² and a population of 5,000, and a sliver that falls within a particular watershed has an area of 3 km², then the estimated population for that sliver is (3/10) × 5,000 = **1,500**.

The result is stored as a `LONG` (integer) field because you cannot have fractional people in a count.

````python
# Cell 13 — Calculate the area-weighted estimated population for each sliver

expression = "(!shape.area! / !{}!) * !{}!".format(iaFieldName, oPopFieldName)
print("Field calculation expression:")
print("  {}".format(expression))

arcpy.management.CalculateField(
    in_table=tintersectFC,
    field=EstPopFieldName,
    expression=expression,
    expression_type="PYTHON3",
    code_block="",
    field_type="LONG"
)

print("\nEstimated population calculated for each sliver.")
print(arcpy.GetMessages())
````

## Preview the Sliver-Level Estimates

Let's look at a few rows from the intersect output to confirm the estimated population values make sense. Each row represents one sliver — a piece of a census tract that falls within a single watershed. The `Estimated_Population` should always be **less than or equal to** the original population (`P0010001`), and the ratio should correspond to the fraction of the tract's area that the sliver represents.

Again, this next code below programmatically goes through the rows and prints values that are important as a verification process.  You could do the same thing by opening the intersect table again and looking at the values for these fields printed below.

````python
# Cell 14 — Preview sliver-level data

preview_fields = [oUnitsfield, oPopFieldName, iaFieldName, "SHAPE@AREA", EstPopFieldName]

print("{:<12} {:>12} {:>16} {:>16} {:>14}".format(
    oUnitsfield, oPopFieldName, iaFieldName, "Sliver_Area", EstPopFieldName))
print("-" * 74)

row_count = 0
with arcpy.da.SearchCursor(tintersectFC, preview_fields) as cursor:
    for row in cursor:
        print("{:<12} {:>12,} {:>16,.2f} {:>16,.2f} {:>14,}".format(
            row[0], row[1], row[2], row[3], row[4]))
        row_count += 1
        if row_count >= 10:
            break

print("\n(Showing first 10 slivers)")
````

## Step 4: Pairwise Dissolve — Reaggregating Slivers into Watersheds

Now that every sliver has an estimated population, we need to **sum those estimates by watershed**. The **Pairwise Dissolve** tool does exactly this: it merges all polygons that share the same value in a specified field (here, `HUC_8`) into a single polygon, and it can simultaneously compute **summary statistics** on other fields.

We specify `Estimated_Population SUM` as the statistics field, which tells the tool to add up all the `Estimated_Population` values for slivers belonging to the same watershed.

The result is one polygon per watershed, with a field called `SUM_Estimated_Population` containing our final population estimate.

**Tool reference:** [Pairwise Dissolve (Analysis)](https://pro.arcgis.com/en/pro-app/latest/tool-reference/analysis/pairwise-dissolve.htm)

````python
# Cell 15 — Dissolve (reaggregate) sliver estimates by watershed

# arcpy.analysis.PairwiseDissolve(in_features, out_feature_class, dissolve_field, statistics_fields, multi_part)
#   in_features      : the intersect result with sliver-level estimates
#   out_feature_class: the final output feature class
#   dissolve_field   : the field to group by (watershed ID)
#   statistics_fields : "FieldName SUM" to sum the estimated population
#   multi_part       : "MULTI_PART" allows dissolved polygons to be multipart

arcpy.analysis.PairwiseDissolve(
    in_features=tintersectFC,
    out_feature_class=disaggoutFC,
    dissolve_field=oUnitsfield,
    statistics_fields=TotalPopFieldName,
    multi_part="MULTI_PART"
)

print("Pairwise Dissolve complete.")
print("Output written to: {}".format(disaggoutFC))
print(arcpy.GetMessages())
````

## Explore the Final Output

Let's add the dissolved output to the map and print its attribute table. You should see one row per watershed, each with a `SUM_Estimated_Population` field. Compare the watershed shapes to the original watershed boundaries — they should match.

If you style this layer with graduated colors on `SUM_Estimated_Population`, you will have a **choropleth map of estimated population by watershed**.

````python
# Cell 16 — Add the final output to the map and preview the results
# Again should be loaded already if using ArcGIS Pro

# The dissolve tool names the sum field with a "SUM_" prefix
sum_field = "SUM_{}".format(EstPopFieldName)

print("{:<12} {:>22}".format(oUnitsfield, sum_field))
print("-" * 36)

with arcpy.da.SearchCursor(disaggoutFC, [oUnitsfield, sum_field]) as cursor:
    for row in cursor:
        print("{:<12} {:>22,}".format(row[0], row[1]))
````

## Quick Accuracy Check: Comparing Total Population

A useful check is to compare the **sum of the original census tract populations** with the **sum of the watershed-level estimates**. Because we intersected these two layers, there will be some "people" (i.e. part of cencus tracts) that fell outside of the watershed boundaries. You can see this if you turn on the ctracts layer and compare it spatial extent to the intersect layer. Also we used integer rounding when calculating the estimated population for each sliver so even if the spatial extents were intitally identical the totals will not match exactly, but they should be very close.

This kind of bookkeeping is an important habit whenever you redistribute data across boundaries.

````python
# Cell 17 — Compare total population before and after disaggregation

sum_field = "SUM_{}".format(EstPopFieldName)

# Sum the original population from the census tracts
original_total = 0
with arcpy.da.SearchCursor(disaggFC, [oPopFieldName]) as cursor:
    for row in cursor:
        original_total += row[0]

# Sum the estimated population from the dissolved watersheds
estimated_total = 0
with arcpy.da.SearchCursor(disaggoutFC, [sum_field]) as cursor:
    for row in cursor:
        estimated_total += row[0]

difference = original_total - estimated_total
pct_diff = (difference / original_total) * 100 if original_total != 0 else 0

print("Original total population (census tracts) : {:>12,}".format(original_total))
print("Estimated total population (watersheds)   : {:>12,}".format(estimated_total))
print("Difference                                : {:>12,}".format(difference))
print("Percent difference                        : {:>11.2f}%".format(pct_diff))

if abs(pct_diff) < 1:
    print("\nThe totals are within 1% — the disaggregation is consistent and the spatial extent of the two inputs likel was simmilar")
else:
    print("\nThe totals differ by more than 1% meaning the difference is the estimated total living outside the watershed boundaries")
````

## Preparing Data for Visualization with Pandas

Now that we have our final population-by-watershed data, let's create a bar chart to visualize the results. To do this, we will:

1. Read the dissolved feature class into a **pandas DataFrame** using a SearchCursor.
2. Sort the data by population.
3. Plot it with **matplotlib**.

Pandas is Python's most popular library for tabular data manipulation. A DataFrame is essentially a table (rows and columns) with powerful methods for sorting, filtering, grouping, and summarizing. Converting our GIS attribute table into a DataFrame gives us access to all of those capabilities.

We only need the `HUC_8` identifier and the summed population field — the geometry is not needed for a chart.

````python
# Cell 18 — Read the dissolved output into a pandas DataFrame

sum_field = "SUM_{}".format(EstPopFieldName)

# Build a list of dictionaries from the SearchCursor, then create a DataFrame
rows = []
with arcpy.da.SearchCursor(disaggoutFC, [oUnitsfield, sum_field]) as cursor:
    for row in cursor:
        rows.append({"HUC_8": row[0], "Estimated_Pop": row[1]})

df = pd.DataFrame(rows)

# Sort by population in descending order for a cleaner chart
df = df.sort_values("Estimated_Pop", ascending=True).reset_index(drop=True)

print("DataFrame created with {} watersheds:\n".format(len(df)))
print(df.to_string(index=False))
````

## Creating a Horizontal Bar Chart of Population by Watershed

A horizontal bar chart is a good choice here because the HUC-8 codes are long labels that display better on the y-axis than rotated along the x-axis. We sort the bars from smallest to largest (bottom to top) so the largest-population watershed appears at the top — a common convention that makes the chart easy to scan.

**Matplotlib** is Python's foundational plotting library. The pattern below — create a figure and axes, plot the data, label the axes, add a title, then show — is one you will use in nearly every visualization you make.

````python
# Cell 19 — Create a horizontal bar chart of population by watershed

fig, ax = plt.subplots(figsize=(10, max(6, len(df) * 0.4)))

ax.barh(df["HUC_8"], df["Estimated_Pop"], color="steelblue", edgecolor="black", linewidth=0.5)

ax.set_xlabel("Estimated Population", fontsize=12)
ax.set_ylabel("HUC-8 Watershed", fontsize=12)
ax.set_title("Estimated Population by HUC-8 Watershed\n(Area-Weighted Disaggregation from Census Tracts)", fontsize=14)

# Add value labels at the end of each bar
for index, value in enumerate(df["Estimated_Pop"]):
    ax.text(value + (df["Estimated_Pop"].max() * 0.01), index, "{:,}".format(value),
            va="center", fontsize=9)

# Add a little extra space on the right for the labels
ax.set_xlim(0, df["Estimated_Pop"].max() * 1.15)

plt.tight_layout()
plt.show()
print("Chart rendered.")
````

## Summary Statistics Table

Finally, let's use pandas to compute some basic descriptive statistics for the estimated population across all watersheds. This gives a quick quantitative sense of the distribution — the minimum, maximum, mean, median, and standard deviation. These numbers complement the bar chart by providing exact values.

````python
# Cell 20 — Summary statistics for the estimated watershed populations

print("Summary Statistics for Estimated Population by Watershed")
print("=" * 55)
print("  Number of watersheds : {:>12}".format(len(df)))
print("  Total population     : {:>12,}".format(df["Estimated_Pop"].sum()))
print("  Minimum              : {:>12,}".format(df["Estimated_Pop"].min()))
print("  Maximum              : {:>12,}".format(df["Estimated_Pop"].max()))
print("  Mean                 : {:>12,.1f}".format(df["Estimated_Pop"].mean()))
print("  Median               : {:>12,.1f}".format(df["Estimated_Pop"].median()))
print("  Std. Deviation       : {:>12,.1f}".format(df["Estimated_Pop"].std()))
````
