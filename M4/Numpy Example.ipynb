```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np

fc = r"C:\temp\LabData.gdb\watersheds"

import arcpy
array = arcpy.da.FeatureClassToNumPyArray(fc, ["*"])

print(array["HUC_8"])

plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
watersheds = array["HUC_8"]
y_pos = np.arange(len(watersheds))
areatot = array["Shape_Area"]
ax.barh(y_pos, areatot, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(watersheds)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Watershed Area')
ax.set_title('Area of the Watershed of South Mississippi')

plt.show()

array
```
