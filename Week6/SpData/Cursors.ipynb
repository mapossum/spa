```python
import arcpy

fc = r'Tornados_Since_1970_MS'
fields =  ["year", "SHAPE@", "mo"] #"*" 


sr = arcpy.SpatialReference(3814)

totlength = 0
c = 0
with arcpy.da.SearchCursor(fc, fields, "year > 2000", sr) as cursor:
    for row in cursor:
        #print(row[1].length)
        totlength += row[1].length
        c += 1
        
print(totlength / c, c)
```

```python
import arcpy

fc = r'Tornados_Since_1970_MS'
fields =  ["year", "mag", "SHAPE@", "mo"] #"*" 

totlength = 0
totmag = 0
c = 0
with arcpy.da.SearchCursor(fc, fields, "year > 2000") as cursor:
    for row in cursor:
        #print(row[1].length)
        mag = row[1]
        if (mag > 0):
            totmag += row[1]
            c += 1
        
        
print(totmag / c, c)
```

```python
import arcpy

fc = r'Tornados_Since_1970_MS'
fields = ['wid', "windkph"]

arcpy.AddField_management(fc, "windkph", "LONG")

#Update Wind to Kph from Mph
cursor = arcpy.da.UpdateCursor(fc, fields)
for row in cursor:
    windmph = row[0]
    windkph = 1.60934 * windmph
    row[1] = windkph
    cursor.updateRow(row)
    
del cursor
```

```python
import arcpy
import numpy

arcpy.env.overwriteOutput = True

outFC = r"pointlocations"

# Create a numpy array with an id field, and a field with a tuple 
#  of x,y coordinates
#
array = numpy.array([(1, 471316.3835861763, 5000448.782036674),
                     (2, 470402.49348005146, 5000049.216449278),
                    (3, 270402.49348005146, 3000049.216449278)],
                    numpy.dtype([('idfield',numpy.int32),('x', numpy.float32),('y', numpy.float32)]))

sr = arcpy.SpatialReference(3814)

# Export the numpy array to a feature class using the XY field to
#  represent the output point feature
#
arcpy.da.NumPyArrayToFeatureClass(array, outFC, ['x', "y"], sr)
```

```python
import arcpy
import numpy as np

fields = ['OBJECTID', "SHAPE@LENGTH", "SHAPE"]
fc = r'Tornados_Since_1970_MS'
outFC = r"Torndado_pointlocations"

array = arcpy.da.FeatureClassToNumPyArray(fc, fields)

sr = arcpy.Describe(fc).spatialReference

#Change the field names that have symbols such as the @ sign
array.dtype.names = ['OBJECTID', "SPLENGTH", "SHAPE"]
arcpy.da.NumPyArrayToFeatureClass(array, outFC, ["SHAPE"], sr)
```

```python
import arcpy
import numpy as np

fields = ['OBJECTID', "SHAPE@LENGTH"]
fc = r'Tornados_Since_1970_MS'
#outFC = r'Tornados_Since_1970_MS_updated'

array = arcpy.da.FeatureClassToNumPyArray(fc, fields)

#Change the field names that have symbols such as the @ sign
array.dtype.names = ['LINKID', "KM_LENGTH" ]

array["KM_LENGTH"] = array["KM_LENGTH"] / 1000

arcpy.da.ExtendTable(fc, "OBJECTID", array, "LINKID")

del array, fc
```
